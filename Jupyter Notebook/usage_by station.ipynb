```python
# create a new column with the difference between the current num_bikes_available and the previous one

data = data.sort_values(by=['station_id', 'load'])
data['usage'] = data['num_bikes_available'].diff()
data[data['station_id'] == 'emobility:69636']
```

```python
# create a new column called 'usage' with the difference between the current num_bikes_available and the previous one
data['usage'] = data.groupby(['station_id'])['num_bikes_available'].diff()*(-1)
columns_list = ['load','station_id', 'num_bikes_available', 'usage']
# Set negative values to 0
data['usage'] = data['usage'].apply(lambda x: max(0, x))
usage_df = data[columns_list]
# Check one single station_id
usage_df[usage_df['station_id'] == 'emobility:69636']
```

```python
def shift_with_latter(available):
    shift = (available.shift() - available).cumsum()
    return shift

def difference_with_latter(available:np.array):
    difference = (available.diff()).cumsum()
    return difference

# Group the data frame by range_identifier and aggregate the variables
new_df = data.groupby(['station_id', 'load']).agg(            
    usage=pd.NamedAgg(column='num_bikes_available', aggfunc=difference_with_latter))
    
new_df.head()

   # (df[cond_col_name].shift() != df[cond_col_name])
```
